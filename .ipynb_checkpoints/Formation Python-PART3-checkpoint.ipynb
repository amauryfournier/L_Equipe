### Feature engineering

```python
df_client_lite.premiere_visite = pd.to_datetime(df_client_lite.premiere_visite, format='%Y-%m-%d %H:%M:%S')
df_client_lite.derniere_visite = pd.to_datetime(df_client_lite.derniere_visite, format='%Y-%m-%d %H:%M:%S')
```

```python
liste_sites = [492987,496306,496307,496838,502193,509043,539121,548647]
df_client_lite = df_client_lite.loc[df_client_lite.id_site.isin(liste_sites)]
df_client_lite.id_site.value_counts()
```

```python
df_client_lite = df_client_lite.loc[df_client_lite.derniere_visite < datetime.date(2015,8,1)]
```

```python
df_client_lite.premiere_visite = df_client_lite.groupby(["id_client"]).premiere_visite.transform(min)
df_client_lite.derniere_visite = df_client_lite.groupby(["id_client"]).derniere_visite.transform(max)
df_client_lite.sort_values(by='id_client')
```

```python
df_client_lite['recence_de_visite'] =(datetime.date(2015, 8, 1) - df_client_lite.derniere_visite)
df_client_lite.recence_de_visite = df_client_lite.recence_de_visite.astype('timedelta64[D]').astype(int)

df_client_lite['anciennete'] =(datetime.date(2015, 8, 1) - df_client_lite.premiere_visite)
df_client_lite.anciennete = df_client_lite.anciennete.astype('timedelta64[D]').astype(int)
df_client_lite.head(1)
```

```python
df_client_lite["max_visites"] = df_client_lite.groupby('id_client').visites.transform('max')
df_client_lite["min_visites"] =  df_client_lite.groupby('id_client').visites.transform('min')
df_client_lite["somme_visites"] =  df_client_lite.groupby('id_client').visites.transform('sum')
df_client_lite["max_pages_vues"] =  df_client_lite.groupby('id_client').pages_vues.transform('max')
df_client_lite["min_pages_vues"] = df_client_lite.groupby('id_client').pages_vues.transform('min')
df_client_lite["somme_pages_vues"] = df_client_lite.groupby('id_client').pages_vues.transform('sum')
df_client_lite['moyenne_pages_vues']  = df_client_lite.groupby('id_client').pages_vues.transform('mean')
```

```python
df_client_lite.drop(['fileid','premiere_visite','derniere_visite','visites',"pages_vues"], axis=1,inplace=True)
df_client_lite = df_client_lite.drop_duplicates()
```

```python
df_client_lite = pd.get_dummies(df_client_lite,columns= ["id_site"])
```

```python
#creation d'une liste contenant les colonnes id_site
#première methode
liste_colonnes = []
for col in df_client_lite.columns :
    if col.startswith("id_site") :
        liste_colonnes.append(col)
liste_colonnes

#seconde methode
l=[col for col in df_client_lite.columns if col.startswith("id_site")]
```

```python
df_client_lite[liste_colonnes] = df_client_lite.groupby('id_client')[liste_colonnes].transform('max')
```

```python
df_client_lite = df_client_lite.drop_duplicates()
```

```python
df_session_lite = df_session_lite.loc[df_session_lite.fin_de_session != u'0001-01-01 00:00:00']
```

```python
#formatage datetime 
df_session_lite.debut_de_session = pd.to_datetime(df_session_lite['debut_de_session'], format='%Y-%m-%d %H:%M:%S')
df_session_lite.fin_de_session = pd.to_datetime(df_session_lite['fin_de_session'], format='%Y-%m-%d %H:%M:%S')
df_session_lite['duree_session'] = df_session_lite['fin_de_session'] - df_session_lite['debut_de_session']
```

```python
df_session_lite.sort_values(by='fin_de_session')
```

```python
df_session_lite.loc[df_session_lite['localisation_3eme_niveau'] == 'Paris', 'Paris'] = 1
df_session_lite.loc[df_session_lite['localisation_3eme_niveau'] != 'Paris', 'Paris'] = 0

df_session_lite.loc[df_session_lite['localisation_2eme_niveau'] == 'Ile-de-France', 'RP'] = 1
df_session_lite.loc[df_session_lite['localisation_2eme_niveau'] != 'Ile-de-France', 'RP'] = 0

df_session_lite.loc[df_session_lite['localisation_1er_niveau'] == 'France', 'France'] = 1
df_session_lite.loc[df_session_lite['localisation_1er_niveau'] != 'France', 'France'] = 0
```

#### Regex

```python
df_session_lite["os2"] = "autres"
df_session_lite.os = df_session_lite.os.str.lower()
df_session_lite.loc[df_session_lite.os.str.match(".*ios.*"), "os2"] = "iphone"
df_session_lite.loc[df_session_lite.os.str.match(".*ipad.*"), "os2"] = "ipad"
df_session_lite.loc[((df_session_lite.os.str.match(".*mac os.*")) | (df_session_lite.os2.str.match(".*os x.*")))  , "os2"] = "mac"
df_session_lite.loc[df_session_lite.os.str.match(".*windows.*"), "os2"] = "windows"
df_session_lite.loc[df_session_lite.os.str.match(".*windows phone.*"), "os2"] = "windows_phone"
df_session_lite.loc[df_session_lite.os.str.match(".*android.*"), "os2"] = "android"
df_session_lite.loc[df_session_lite.os.str.match(".*linux.*"), "os2"] = "linux"
df_session_lite.loc[df_session_lite.os.str.match(".*chrome.*"), "os2"] = "chrome_os"
```
